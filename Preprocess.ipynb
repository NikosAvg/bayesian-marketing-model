```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
# Read The Raw Data
raw_data = 'Data/Raw/marketing_campaign.csv'
clean_data = 'Data/Clean/marketing_campaign.csv'
df = pd.read_csv(raw_data, sep=';')
df['Age'] = 2025 - df['Year_Birth'] # Turn Year of Birth To Age
df.head()
```

```python
# Drop NA and focus on certain features
df = df[['Age', 'Income', 'Education', 'Response']].dropna()
```

```python
# Limit our education levels to 3
df = df[df['Education'].isin(['Basic', 'Graduation', 'Master'])]
```

```python
# One-hot encode education
df = pd.get_dummies(df, columns=['Education'], drop_first=True)
df.to_csv(clean_data, sep=';', index=False)
```
